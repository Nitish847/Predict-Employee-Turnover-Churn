```python
y=df1.left
y
```

```python
X=df1.drop("left",axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.20,random_state=0,stratify=y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score
acc_score = [] #to store accuracy values for different paraemters
for M,N in zip(range(12),range(12)):
    M=M+2
    N=N+2
   
    model1 = RandomForestClassifier(n_estimators = 120,max_leaf_nodes=M,min_samples_split=N,random_state=0)

    model1.fit(X_train, y_train)  #fit the model
    pred=model1.predict(X_test) #make prediction on test set
    score = accuracy_score(y_test,pred) #calculate accuracy
    acc_score.append(score) #store accuracy values
    print("Accuracy value for M = {} and N= {} IS {}".format(M,N,score))
```

```python
model2 = RandomForestClassifier(n_estimators = 120,max_leaf_nodes=13,min_samples_split=13,random_state=0)
```

```python
model2.fit(X_train,y_train)
```

```python
Predictions=model2.predict(X_test)
Predictions
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Predictions,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Predictions,y_test)
```

```python
from sklearn.metrics import precision_score
precision_score(Predictions,y_test)
```

```python
from sklearn.metrics import recall_score
recall_score(Predictions,y_test)
```
